Python Functions

A function in Python is a reusable block of code that performs a specific task. Functions help in structuring the code, making it modular, readable, and reusable.

Types of Functions in Python:

    1.Built-in Functions: Python provides many built-in functions like print(), len(), sum(), etc.
    2.User-defined Functions: Functions created by the user to perform specific tasks.
    3.Lambda Functions: Anonymous functions created using the lambda keyword.
    4.Recursive Functions: Functions that call themselves.

**Map Function**

The map() function in Python is used to apply a given function to all items in an iterable (like a list, tuple, or set) and return an iterator with the results.

Syntax:

map(function, iterable)

function: A function that will be applied to each item in the iterable.
iterable: The iterable whose elements will be processed.



```python
#Using map() with a Built-in Function
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(pow, numbers, [2]*len(numbers))  # Squares each number
print(list(squared_numbers))
```

```python
#Using map() with a Lambda Function
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x**2, numbers)
print(list(squared_numbers))
```

```python
#Using map() with Multiple Iterables
a = [1, 2, 3]
b = [4, 5, 6]
sum_list = map(lambda x, y: x + y, a, b)
print(list(sum_list))
```

```python
#Using map() with a User-defined Function
def double(num):
    return num * 2

numbers = [1, 2, 3, 4, 5]
doubled_numbers = map(double, numbers)
print(list(doubled_numbers))
```

**filter() Function in Python**

The filter() function in Python is used to filter elements from an iterable based on a condition provided by a function. It returns an iterator containing only the elements for which the function evaluates to True.

Syntax:

*filter(function, iterable)*

function: A function that returns True or False for each element.
iterable: The iterable (list, tuple, set, etc.) to be filtered.




```python
#Using filter() with a Built-in Function
numbers = [-5, -2, 0, 3, 7, 10]
positive_numbers = filter(lambda x: x > 0, numbers)
print(list(positive_numbers))
```

```python
#Using filter() with a User-defined Function
def is_even(n):
    return n % 2 == 0

numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(is_even, numbers)
print(list(even_numbers))
```

```python
#Filtering Strings

#Filtering out words that start with 'A':

words = ["Apple", "Banana", "Avocado", "Grapes"]
filtered_words = filter(lambda word: word.startswith("A"), words)
print(list(filtered_words))
```

```python
#Filtering None or Empty Strings
items = ["Python", "", None, "Java", "C++", ""]
non_empty_items = filter(None, items)
print(list(non_empty_items))

#Here, filter(None, iterable) removes Falsey values (None, "", 0, False).
```

**reduce() Function in Python**

The reduce() function is used to apply a function cumulatively to the elements of an iterable, reducing it to a single value. It is part of the functools module.

Syntax

from functools import reduce

reduce(function, iterable[, initializer])



*   function: A function that takes two arguments and performs an operation
* iterable: The iterable (list, tuple, etc.) to reduce.
* initializer (optional): A starting value.


```python
#Sum of Elements (Basic Example)
from functools import reduce

numbers = [1, 2, 3, 4, 5]
result = reduce(lambda x, y: x + y, numbers)
print(result)
```

```python
#Finding the Maximum Value
numbers = [3, 8, 1, 6, 10, 2]
max_value = reduce(lambda x, y: x if x > y else y, numbers)
print(max_value)
```

```python
#Using reduce() with an Initial Value

#If we provide an initializer, it starts with that value:

numbers = [1, 2, 3, 4, 5]
result = reduce(lambda x, y: x * y, numbers, 10)
print(result)


# How it works:

#     Step 1: 10 * 1 = 10
#     Step 2: 10 * 2 = 20
#     Step 3: 20 * 3 = 60
#     Step 4: 60 * 4 = 240
#     Step 5: 240 * 5 = 1200
```

```python
#Concatenating Strings
words = ["Hello", " ", "World", "!"]
sentence = reduce(lambda x, y: x + y, words)
print(sentence)
```

#Use Cases

**Extract domain names from email addresses**




```python
emails = ["user1@gmail.com", "admin@yahoo.com", "test@outlook.com"]
domains = list(map(lambda email: email.split('@')[1], emails))
print(domains)
```

**Use Case: Extracting Hashtags from Posts**

```python
posts = [
    "AI is transforming the world! #AI #MachineLearning",
    "Python is amazing for data science! #Python #DataScience",
    "Climate change is real. #Environment #ClimateAction",
]

# Step 1: Extract hashtags using map() and filter()
hashtags = list(map(lambda p: [word for word in p.split() if word.startswith("#")], posts))

# Step 2: Flatten the list
flattened_hashtags = reduce(lambda x, y: x + y, hashtags)

print(flattened_hashtags)
```

**Use Case: Identifying Patients at Risk**

```python
patients = [
    {"id": 1, "name": "John", "heart_rate": 72},
    {"id": 2, "name": "Emma", "heart_rate": 110},
    {"id": 3, "name": "Liam", "heart_rate": 98},
    {"id": 4, "name": "Sophia", "heart_rate": 105},
]

# Step 1: Filter patients with high heart rates
at_risk_patients = list(filter(lambda p: p["heart_rate"] > 100, patients))

print(at_risk_patients)
```

#Local and Global Variables in Python

In Python, variables have a scope, meaning where they can be accessed and modified. Variables can be classified as local or global, depending on where they are declared.

1. Global Variables

    Defined outside of functions or classes.
    Accessible throughout the entire script.
    Can be modified inside functions using the global keyword.

Example: Using a Global Variable

```python
x = 10  # Global variable

def display():
    print("Global x:", x)  # Accessing the global variable

display()
print("Outside function, x:", x)
```

 2. Local Variables

    Declared inside a function.
    Accessible only within that function.
    Not available outside the function.

Example: Using a Local Variable

```python
def test():
    y = 20  # Local variable
    print("Local y:", y)

test()
print("Outside function, y:", y)  # This will cause an error
```

3. Modifying Global Variables Inside Functions

If you want to modify a global variable inside a function, you need to use the **global** keyword.

Example: Changing a Global Variable

```python
z = 100  # Global variable

def modify():
    global z  # Declaring that we are using the global variable
    z = 200   # Modifying the global variable
    print("Inside function, z:", z)

modify()
print("Outside function, z:", z)  # z is now updated globally
```

4. Nested Functions and Nonlocal Variables

    Variables in nested functions can use the nonlocal keyword.
    nonlocal allows modifying variables from an enclosing function (but not global variables).

Example: Using nonlocal in a Nested Function

```python
def outer():
    a = 5  # Local variable in outer function

    def inner():
        nonlocal a  # Refers to 'a' in the outer function
        a = 10  # Modifies 'a' in the outer function
        print("Inside inner function, a:", a)

    inner()
    print("Inside outer function, a:", a)

outer()
```
